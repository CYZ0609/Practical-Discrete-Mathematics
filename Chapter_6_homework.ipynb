```
# Import necessary libraries
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# Step 1: Download the image
image_url = "https://upload.wikimedia.org/wikipedia/commons/e/e0/2014_Rallye_Deutschland_by_2eight_8SC0443.jpg"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))

# Step 2: Resize the image to double its size
original_size = image.size
new_size = (original_size[0] * 2, original_size[1] * 2)
resized_image = image.resize(new_size, Image.Resampling.LANCZOS)

# Step 3: Display the original and resized images
plt.figure(figsize=(12, 6))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title(f"Original Image\nSize: {original_size}")
plt.axis("off")

# Display resized image
plt.subplot(1, 2, 2)
plt.imshow(resized_image)
plt.title(f"Resized Image\nSize: {new_size}")
plt.axis("off")

# Show both images
plt.show()
```

```
# Import necessary libraries
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# Step 1: Download the image
image_url = "https://upload.wikimedia.org/wikipedia/commons/e/e0/2014_Rallye_Deutschland_by_2eight_8SC0443.jpg"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))

# Step 2: Transpose the image by 90 degrees
transposed_image = image.transpose(Image.ROTATE_90)

# Step 3: Display the original and transposed images
plt.figure(figsize=(12, 6))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Display transposed image
plt.subplot(1, 2, 2)
plt.imshow(transposed_image)
plt.title("Transposed Image (90°)")
plt.axis("off")

# Show both images
plt.show()

# Step 4: Save the transposed image if needed
transposed_image.save("transposed_image.jpg")
print("Transposed image saved as 'transposed_image.jpg'")
```

```
!pip install pillow

from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# Load the images
url1 = 'https://upload.wikimedia.org/wikipedia/commons/thumb/e/e0/2014_Rallye_Deutschland_by_2eight_8SC0443.jpg/800px-2014_Rallye_Deutschland_by_2eight_8SC0443.jpg'  # First image URL
url2 = 'https://upload.wikimedia.org/wikipedia/commons/thumb/8/8d/The_Empire_Hotel_06.jpg/1280px-The_Empire_Hotel_06.jpg'  # Second image URL

response1 = requests.get(url1)
img1 = Image.open(BytesIO(response1.content))

response2 = requests.get(url2)
img2 = Image.open(BytesIO(response2.content))

# Display the original images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.title('Image 2')
plt.axis('off')

plt.show()

# Combine the images vertically
width = max(img1.width, img2.width)
total_height = img1.height + img2.height

combined_img = Image.new('RGB', (width, total_height))
combined_img.paste(img1, (0, 0))
combined_img.paste(img2, (0, img1.height))

# Display the combined image
plt.imshow(combined_img)
plt.title('Combined Image')
plt.axis('off')
plt.show()
```

```
# Import necessary libraries
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# Step 1: Define the URL of the image
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/e0/2014_Rallye_Deutschland_by_2eight_8SC0443.jpg/800px-2014_Rallye_Deutschland_by_2eight_8SC0443.jpg"

# Step 2: Load the image from the URL
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))

# Step 3: Convert the image to grayscale
gray_image = image.convert("L")  # 'L' mode is for grayscale

# Step 4: Display the grayscale image
plt.figure(figsize=(8, 8))
plt.imshow(gray_image, cmap='gray')  # Use 'gray' colormap for proper grayscale display
plt.title("Grayscale Image")
plt.axis("off")  # Hide axes for better viewing
plt.show()

# Optional: Save the grayscale image
gray_image.save("grayscale_image.jpg", "JPEG")
print("Grayscale image saved as 'grayscale_image.jpg'")
```
